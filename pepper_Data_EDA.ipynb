```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("silver_df.csv")
```

```python
df.head()
```

```python
df.tail()
```

drop the last 4 rows

```python
df.drop(index=df.index[-4 :],axis=0,inplace=True)
```

```python
df.tail()
```

```python
df.info()
```

```python
df['week_end_dt']= pd.to_datetime(df['week_end_dt'])
```

```python
df['year']= df['week_end_dt'].dt.year
```

```python
df['month']= df['week_end_dt'].dt.month
```

```python
# define new column to find the percentage of each country sypply from the total volume
column= ['brazil','china', 'india', 'indonesia', 'vietnam']
for col in column:
  df[f'{col}_per_vol']= (df[col]/ df['total_volume'] ) *100
```

```python
df
```

try to find if there is a relation between the price going out of the min max range and the supply demand differences

```python
df['in_range']= df['price']*0
```

```python
len(df['price'])
```

```python
df_new= df.copy()
```

```python
for i in range(0,len(df['price'])):
  if (df_new['price'][i] >= df_new['jordan_min_price'][i])& (df_new['price'][i]<= df_new['jordan_max_price'][i]):
    df_new['in_range'][i]= 0
  else:
    df_new['in_range'][i]= 1
```

```python
df_new['ratio']=  df['supply']/df['demand']
```

```python
df_new['per']= df['total_volume']/df['supply']
```

```python
print(df_new[['jordan_min_price', 'jordan_max_price','per', 'ratio', 'in_range', 'price']][500:600].to_string())
```

```python
correl=df_new[['jordan_min_price', 'jordan_max_price','per', 'ratio', 'in_range', 'price']].corr()
```

```python
sns.heatmap(correl,cbar= True, square= True, fmt= '.1f' ,annot= True, annot_kws= {'size':8}, cmap='Blues')
```
